```python
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.dpi'] = 200
```

```python
from math import pi
```

```python
N = 50
x = range(1, N+2)
y = [4 * sum([(-1) ** k / (2*k + 1) for k in range(n)]) for n in x]

plt.axhline(pi, linestyle="--", color="tab:orange", label=r"$\pi$")
plt.plot(range(N+1), y, label=r"$4\sum_{k=0}^n \frac{(-1)^k}{(2k+1)}$")

plt.xlabel("n")
plt.title(r"Approximating $\pi$")
plt.xlim([0, N])
plt.legend()
```
